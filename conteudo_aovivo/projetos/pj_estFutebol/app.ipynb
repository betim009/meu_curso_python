## 🏆 Projeto 1: Estatísticas de Futebol — JOGADORES

### 🎯 Objetivo do Projeto

A ideia deste projeto é **simular um sistema de estatísticas de futebol**, como se estivéssemos criando uma pequena central de dados de jogadores, parecida com sites que mostram desempenho de atletas (como SofaScore, OneFootball, etc.).

Você vai representar os dados de **jogadores de futebol usando dicionários e listas em Python**. Depois, vai criar **funções para analisar esses dados**, como a quantidade de gols, média de assistências e outras estatísticas importantes.

Por fim, vai desenvolver **uma interface simples no terminal (linha de comando)** para permitir que o usuário escolha o que quer visualizar ou analisar.

---

### 🧱 Estrutura dos Dados

Você vai criar uma base de dados simples com jogadores e suas estatísticas.

Cada jogador terá:

* `id` → número de identificação
* `nome` → nome do jogador
* `gols` → lista de dicionários com informações dos gols:

  * minuto do gol
  * time adversário
  * data do jogo
* `assistencias` → lista de dicionários com assistências:

  * minuto da assistência
  * time adversário
  * data do jogo
* `partidas` → quantidade total de jogos
* `part_direta` → quantidade de participações diretas em gols (gols + assistências)

💡 *Exemplo de estrutura para um jogador:*

```python
dembele = {
    "id": 10,
    "nome": "Dembelé",
    "partidas": 20,
    "part_direta": 8,
    "gols": [
        {"id": 1, "time": "Lyon", "minuto": 45, "data": "01/01/2025"},
        {"id": 2, "time": "Lyon", "minuto": 65, "data": "01/01/2025"},
        {"id": 3, "time": "Olympique", "minuto": 5, "data": "06/01/2025"},
    ],
    "assistencias": [
        {"id": 1, "time": "Lyon", "minuto": 4, "data": "01/01/2025"},
        {"id": 2, "time": "Lyon", "minuto": 15, "data": "01/01/2025"},
        {"id": 3, "time": "Olympique", "minuto": 5, "data": "06/01/2025"},
    ]
}
```

### 🧐 Funções que você vai criar

Você deve implementar algumas funções que vão fazer análises com esses dados:

* ✅ Calcular a **quantidade total de gols** de cada jogador
* ✅ Descobrir **contra qual time o jogador mais marcou gols**
* ✅ Ver **quem tem a melhor média de gols por partida**
* ✅ Ver **quem tem a melhor média de assistências por partida**
* ✅ Ver **quem tem mais participações diretas em gols** (gols + assistências)
* ✅ Calcular a **média de participação direta por partida**

---

### 💻 Interface no terminal

Você vai criar um **menu simples no terminal** com opções, como por exemplo:

```
[1] Ver total de gols por jogador
[2] Ver o time que cada jogador mais marcou gols
[3] Ver jogador com melhor média de gols
[4] Ver jogador com mais assistências
[5] Ver jogador com maior participação direta em gols
[6] Sair
```

O usuário digita o número da opção e o programa executa a função correspondente.
